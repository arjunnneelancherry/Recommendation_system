```python
data=pd.read_csv('book (1).csv',encoding='Latin-1',error_bad_lines=False)
data
```

```python
data=data.sort_values(['User.ID'])
```

```python
data.head()
```

```python
len(data['Book.Title'].unique())
```

```python
data_pivot=data.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
data_pivot
```

```python
data_pivot.index=data['User.ID'].unique()
data_pivot
```

```python
data_pivot.fillna(0,inplace=True)
data_pivot
```

```python
user_sim=1-pairwise_distances(data_pivot,metric='cosine')
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.index=data['User.ID'].unique()
user_sim_df.columns=data['User.ID'].unique()
```

```python
user_sim_df.head()
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df
```

```python
user_sim_df.idxmax(axis=1)
```
